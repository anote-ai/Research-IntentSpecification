```python
# Bad implementation - sorts the output
def bad_solution(numbers):
    mi, ma = min(numbers), max(numbers)
    result = [(x - mi) / (ma - mi) for x in numbers]
    return sorted(result)  # destroys original order

print(bad_solution([2.0, 1.0, 5.0, 3.0, 4.0]))
# C3 should fail - returns [0.0, 0.25, 0.5, 0.75, 1.0] instead of [0.25, 0.0, 1.0, 0.5, 0.75]
```

```python
def bad_solution(numbers):
    # finds closest but doesn't sort output
    min_diff = float('inf')
    pair = None
    for i in range(len(numbers)):
        for j in range(i+1, len(numbers)):
            diff = abs(numbers[i] - numbers[j])
            if diff < min_diff:
                min_diff = diff
                pair = (numbers[i], numbers[j])
    return pair  # doesn't guarantee ascending order

print(bad_solution([1.5, 1.0, 2.1, 1.25, 3.0]))  # C2 should fail - might return (1.5, 1.25)
```

```python
{
    "task_id": "HumanEval/26",
    "spec_type": "algorithm",
    "ambiguous_prompt": "Write a function named `solution` that takes a list of integers and removes any elements that appear more than once.",
    "gold_prompt": "Write a function named `solution` that takes a list of integers, removes all elements that occur more than once, preserves the original order of remaining elements, and returns a new list without mutating the input.",
    "constraints": [
        {
            "id": "C1",
            "description": "Elements appearing more than once are removed",
            "test_code": "result = solution([1, 2, 3, 2, 4])\nassert 2 not in result, f'Duplicate element 2 should be removed, got {result}'"
        },
        {
            "id": "C2",
            "description": "Original order of remaining elements is preserved",
            "test_code": "result = solution([4, 1, 3, 1, 2])\nassert result == [4, 3, 2], f'Expected [4, 3, 2] (order preserved), got {result}'"
        },
        {
            "id": "C3",
            "description": "Input list is not mutated",
            "test_code": "inp = [1, 2, 3, 2, 4]\n_ = solution(inp)\nassert inp == [1, 2, 3, 2, 4], f'Input was mutated: {inp}'"
        },
        {
            "id": "C4",
            "description": "Empty list returns empty list",
            "test_code": "result = solution([])\nassert result == [], f'Expected [], got {result}'"
        }
    ],
    "stated_test_ids": ["C1"],
    "hidden_test_ids": ["C2", "C3", "C4"]
}
```

```python
# Naive implementation that might miss order preservation
def solution(numbers):
    from collections import Counter
    counts = Counter(numbers)
    return [x for x in set(numbers) if counts[x] == 1]  # set() destroys order

# C1 should PASS
result = solution([1, 2, 3, 2, 4])
print(f"C1: {result}, 2 not in result: {2 not in result}")

# C2 should FAIL - set() doesn't preserve order
result = solution([4, 1, 3, 1, 2])
print(f"C2: {result}, expected [4, 3, 2]")
```

```python
# Naive implementation that might get C2 or C3 wrong
def solution(n):
    factors = []
    d = 2
    while d * d <= n:
        while n % d == 0:
            factors.append(d)
            n //= d
        d += 1
    if n > 1:
        factors.append(n)
    return factors  # this is actually correct - testing against a bad one

# Test all constraints
print("C1:", solution(13))  # should be [13]
print("C2:", solution(15))  # should be [3, 5]
print("C3:", solution(8))   # should be [2, 2, 2]

import math
result = solution(70)
print("C4:", result, "product:", math.prod(result))  # should be [2, 5, 7], product 70
```

```python
# Bad implementation - returns unique factors only, wrong order
def bad_solution(n):
    factors = set()
    d = 2
    while d * d <= n:
        while n % d == 0:
            factors.add(d)
            n //= d
        d += 1
    if n > 1:
        factors.add(n)
    return sorted(factors)  # loses repetition

print("\nBad solution:")
print("C3 should fail:", bad_solution(8))  # returns [2] not [2,2,2]
```
