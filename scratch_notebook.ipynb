```python
import json
from evalplus.data import get_human_eval_plus

dataset = get_human_eval_plus()

# point this at your actual spec file
specs_path = "specs.jsonl"
out_path = "canonicals.jsonl"

task_ids = []
with open(specs_path) as f:
    for line in f:
        line = line.strip()
        if line:
            task_ids.append(json.loads(line)["task_id"])

print(f"Found {len(task_ids)} task_ids")

written, missing = 0, []
with open(out_path, "w") as out:
    for tid in task_ids:
        if tid not in dataset:
            missing.append(tid)
            continue
        e = dataset[tid]
        out.write(json.dumps({
            "task_id": tid,
            "entry_point": e["entry_point"],
            "prompt": e["prompt"],
            "canonical_solution": e["canonical_solution"],
            "full_source": e["prompt"] + e["canonical_solution"],
        }) + "\n")
        written += 1

print(f"Wrote {written} canonicals to {out_path}")
if missing:
    print(f"Missing: {missing}")
```

```python
# Naive implementation that might miss order preservation
def solution(numbers):
    from collections import Counter
    counts = Counter(numbers)
    return [x for x in set(numbers) if counts[x] == 1]  # set() destroys order

# C1 should PASS
result = solution([1, 2, 3, 2, 4])
print(f"C1: {result}, 2 not in result: {2 not in result}")

# C2 should FAIL - set() doesn't preserve order
result = solution([4, 1, 3, 1, 2])
print(f"C2: {result}, expected [4, 3, 2]")
```

```python
# Naive implementation that might get C2 or C3 wrong
def solution(n):
    factors = []
    d = 2
    while d * d <= n:
        while n % d == 0:
            factors.append(d)
            n //= d
        d += 1
    if n > 1:
        factors.append(n)
    return factors  # this is actually correct - testing against a bad one

# Test all constraints
print("C1:", solution(13))  # should be [13]
print("C2:", solution(15))  # should be [3, 5]
print("C3:", solution(8))   # should be [2, 2, 2]

import math
result = solution(70)
print("C4:", result, "product:", math.prod(result))  # should be [2, 5, 7], product 70
```

```python
# Bad implementation - returns unique factors only, wrong order
def bad_solution(n):
    factors = set()
    d = 2
    while d * d <= n:
        while n % d == 0:
            factors.add(d)
            n //= d
        d += 1
    if n > 1:
        factors.add(n)
    return sorted(factors)  # loses repetition

print("\nBad solution:")
print("C3 should fail:", bad_solution(8))  # returns [2] not [2,2,2]
```
